```python
import matplotlib.pyplot as plt
%matplotlib inline

from mpl_toolkits.mplot3d import Axes3D

import numpy as np

import pickle as pickle

import sys
path = '../../scripts/'
sys.path.insert(0,path)
from ripser_to_dict import *
from plot_persistence import *
```

```python
with open('params.pickle','rb') as f:
    params = pickle.load(f)
```

```python
def uniform_wiggle(persistence,params):
    
    """Wiggle each point as determined by unit_tot_wiggle
    in a random direction, i.e. each point is moved a little distance
    so that the sum of these distances = unit_tot_wiggle * max_persistence
    (for points at max_persistence it may be a bit smaller)"""
    
    # extract relevant parameters
    all_vertices_born_at_0 = params['all_vertices_born_at_0']
    n_points = persistence['nb_neuron'] # Find a suitable way to get the number of points, ptobably by adding them to the persistence dictionaries in the long run
    max_persistence = params['ripser']['max_persistence']
    unit_tot_wiggle = params['unit_tot_wiggle']
    wiggle_seed = params['wiggle_seed']
    all_dims = range(params['ripser']['dim'] + 1)
    
    tot_wiggle = unit_tot_wiggle * max_persistence
    delta = tot_wiggle / n_points
    
    np.random.seed(wiggle_seed)
    
    # create new wiggled dictionary
    new_persistence = persistence.copy()
    for dim in all_dims:
        if dim in persistence:
            for point, position in enumerate(persistence[dim].copy()):
                new_position = [-1,-1]
                while not 0 <= new_position[0] <= new_position[1] <= max_persistence:
                    if all_vertices_born_at_0 and dim == 0:
                        p = 0 # we may not want to move dim = 0 points away from birth = 0
                    else:
                        p = 1 - 2 * np.random.rand() # otherwise we create a random value in [0,1]
                    plusminus = 1 - 2 * np.random.randint(2)
                    q = plusminus * np.sqrt(1 - p**2) # we create another random value so that p^2+q^2=1
                    new_position = position + delta * np.array([p,q])
                    # the value is potentially infinite, so we want to keep that feature
                    if position[1] == max_persistence:
                        new_position[1] = max_persistence
                        if position[0] == max_persistence:
                            new_position[0] == max_persistencs
                new_persistence[dim][point] = new_position
                #print('old: ' + str(position[0]) + ', ' + str(position[1]) + ', and new: ' + str(new_position[0]) + ', ' + str(new_position[1]))
    return new_persistence
```

```python
# Test
shape = 'mickey_mouse'
for seed in [0,1,2,3,4]:
    code_to_data = params['paths']['code_to_data'] + 'persistences/'\
    + shape + '/' + str(seed) + '.pickle'
    with open(code_to_data, 'rb') as f:
        persistence = pickle.load(f)
    fig0 = plot_persistence(persistence,params,title='original')
    new_persistence = uniform_wiggle(persistence,params)
    fig1 = plot_persistence(new_persistence,params,title='wiggled')
```
