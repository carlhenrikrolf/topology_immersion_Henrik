# Random points on a sphere

```python
# load packages
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
```

```python
# parameters
n_points = 200
seed = 11
error = 0.1 # the thickness of the noisy sphere (which is bounded by râ‰¤1).
```

## Perfect sphere

```python
# construct perfect (unit) sphere in spherical coordinates: r=1, 0<theta<pi, 0<phi<2*pi
np.random.seed(seed=seed)
perfect_spherical_coords = np.random.rand(3,n_points)
perfect_spherical_coords[0,:] = np.ones((1,n_points))
perfect_spherical_coords[1,:] = perfect_spherical_coords[1,:]*np.pi
perfect_spherical_coords[2,:] = perfect_spherical_coords[2,:]*2*np.pi
```

```python
# transform perfect sphere into Euclidean coordinates
perfect_Euclidean_coords = np.zeros((3,n_points))
perfect_Euclidean_coords[0,:] = perfect_spherical_coords[0,:]*np.sin(perfect_spherical_coords[1,:])*np.cos(perfect_spherical_coords[2,:])
perfect_Euclidean_coords[1,:] = perfect_spherical_coords[0,:]*np.sin(perfect_spherical_coords[1,:])*np.sin(perfect_spherical_coords[2,:])
perfect_Euclidean_coords[2,:] = perfect_spherical_coords[0,:]*np.cos(perfect_spherical_coords[1,:])
```

```python
# double-check that it is indeed a sphere
def doublecheck_shape(Euclidean_coords):
    fig = plt.figure()
    ax = fig.add_subplot(111, projection='3d')
    ax.scatter(Euclidean_coords[0,:], Euclidean_coords[1,:], Euclidean_coords[2,:])

doublecheck_shape(perfect_Euclidean_coords)
```

```python
# create distance matrix for the perfect sphere
perfect_pairwise_distances = np.zeros((n_points,n_points))
for i in range(n_points):
    for j in range(n_points):
        perfect_pairwise_distances[i,j] = np.sqrt((perfect_Euclidean_coords[0,j] - perfect_Euclidean_coords[0,i])**2 + (perfect_Euclidean_coords[1,j] - perfect_Euclidean_coords[1,i])**2 + (perfect_Euclidean_coords[2,j] - perfect_Euclidean_coords[2,i])**2)
```

```python
# export to file
np.savetxt("perfect_sphere_200x200_distance_matrix.txt", perfect_pairwise_distances, delimiter=",")
```

## Noisy sphere

```python
# construct perfect (unit) sphere in spherical coordinates: r=1, 0<theta<pi, 0<phi<2*pi
np.random.seed(seed=seed)
noisy_spherical_coords = np.random.rand(3,n_points)
noisy_spherical_coords[0,:] = 1 - error + error*noisy_spherical_coords[0,:]
noisy_spherical_coords[1,:] = noisy_spherical_coords[1,:]*np.pi
noisy_spherical_coords[2,:] = noisy_spherical_coords[2,:]*2*np.pi
```

```python
# transform perfect sphere into Euclidean coordinates
noisy_Euclidean_coords = np.zeros((3,n_points))
noisy_Euclidean_coords[0,:] = noisy_spherical_coords[0,:]*np.sin(noisy_spherical_coords[1,:])*np.cos(noisy_spherical_coords[2,:])
noisy_Euclidean_coords[1,:] = noisy_spherical_coords[0,:]*np.sin(noisy_spherical_coords[1,:])*np.sin(noisy_spherical_coords[2,:])
noisy_Euclidean_coords[2,:] = noisy_spherical_coords[0,:]*np.cos(noisy_spherical_coords[1,:])
```

```python
# create distance matrix for the perfect sphere
noisy_pairwise_distances = np.zeros((n_points,n_points))
for i in range(n_points):
    for j in range(n_points):
        noisy_pairwise_distances[i,j] = np.sqrt((noisy_Euclidean_coords[0,j] - noisy_Euclidean_coords[0,i])**2 + (noisy_Euclidean_coords[1,j] - noisy_Euclidean_coords[1,i])**2 + (noisy_Euclidean_coords[2,j] - noisy_Euclidean_coords[2,i])**2)
```

```python
# double-check
doublecheck_shape(noisy_Euclidean_coords)
```

```python
# export to file
np.savetxt("noisy_sphere_200x200_distance_matrix.txt", perfect_pairwise_distances, delimiter=",")
```
